```python
rent=pd.read_csv("LA_rent_index.csv")
rent.head()
```

```python
#Rent index for LA
LA_rent_train=[]
for i in range(29,101):
    LA_rent_train.append(int(''.join(c for c in rent.iloc[1][i] if c.isdigit())))
```

```python
df['Issue Date']=pd.to_datetime(df['Issue Date'])
```

```python
permit_type=df["Permit Type"].unique()
permit_train=[]

for year in range(2013,2019):
    for month in range(1,13):
        if month!=12:
            temp=df[(df['Issue Date']>=datetime(year,month,1)) & (df['Issue Date']<datetime(year,month+1,1))]["Permit Type"].value_counts()
        else:
            temp=df[(df['Issue Date']>=datetime(year,month,1)) & (df['Issue Date']<datetime(year+1,1,1))]["Permit Type"].value_counts()
        row=[]
        for permit in permit_type:
            if permit in temp:
                row.append(temp[permit])
            else:
                row.append(0)
        permit_train.append(row)
        
temp=df[(df['Issue Date']>=datetime(2019,1,1)) & (df['Issue Date']<datetime(2019,2,1))]["Permit Type"].value_counts()
permit_JAN2019=[]
for permit in permit_type:
    if permit in temp:
        permit_JAN2019.append(temp[permit])
    else:
        permit_JAN2019.append(0)
```

```python
reg = LinearRegression().fit(permit_train, LA_rent_train)
reg.score(permit_train, LA_rent_train)
```

```python
print ("Predicted JAN 2019 LA rent index: {:.0f}".format(*reg.predict([permit_JAN2019])))
```
